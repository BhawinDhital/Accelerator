```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# 130 - 500 MeV, third harmonics
# Main cavity,97.7 MHz
w_rf = 2*np.pi*97776985.435474600
V_m = 416.25
phi_s = 90*np.pi/180
T = 5.11368e-09 # T/2 actually
N = 105
t = np.linspace(-T,T,N)
A_mc = V_m*np.sin(w_rf*t+phi_s)
plt.xlabel('Time from -T/2 to + T/2 (s)')
plt.ylabel('MV')
plt.title('Main accelerating cavity voltage (MV) versus Time(t)',color='blue')
plt.plot(t,A_mc,color='orange',lw=3,ls='-',label='Main Ac Cavity')
plt.legend(loc=1)
```

```python
# Main cavity + Harmonic cavity
w_rf = 2*np.pi*97776985.435474600
V_h =-46.25
w_th = 3*w_rf
phi_h = 90*np.pi/180
T = 5.11368e-09
N = 105
t = np.linspace(-T,T,N)
B_th = V_h*np.sin(w_th*t+phi_h)
plt.xlabel('Time from -T/2 to + T/2 (s)')
plt.ylabel('MV')
plt.title('Main cavity(orange) and third harmonic cavity(blue) voltage (MV) versus Time(t)',color='blue')
plt.plot(t,A_mc,color='orange',lw=3,ls='-',label='Main_cavity')
plt.plot(t,B_th,color='blue',lw=3,ls='-',label='Third_harmonic')
plt.legend(loc=1)
```

```python
# Main + Harmonic + Bunching cavity
w_rf = 2*np.pi*97776985.435474600
V_b =0.08
phi_b = 180*np.pi/180
T = 5.11368e-09
N = 105
t = np.linspace(-T,T,N)
C_bc = V_b*np.sin(w_rf*t+phi_b)
plt.xlabel('Time from -T/2 to + T/2 (s)')
plt.ylabel('MV')
plt.title('Bunching cavity voltage (MV) versus Time(t)',color='blue')
plt.plot(t,A_mc,color='orange',lw=3,ls='-',label='Main_cavity')
plt.plot(t,B_th,color='blue',lw=3,ls='-',label='Third_harmonic')
plt.plot(t,C_bc,color='red',lw=3,ls='-',label='Bunching_cavity')
plt.legend(loc=1)
```

```python
# Main cavity + Third Harmonic cavity
Main_plus_harmonic = A_mc + B_th
plt.xlabel('Time from -T/2 to + T/2 (s)')
plt.ylabel('MV')
plt.title('Main + Third_harmonic (MV) versus Time(t)',color='blue')
plt.plot(t,Main_plus_harmonic,color='green',lw=3,ls='-',label='Main+Third_harmonic')
plt.plot(t,C_bc,color='red',lw=3,ls='-',label='Bunching_cavity')
plt.legend(loc=4)
```

```python
# Bunch length plot
c = 3e08
z= -c*t
plt.xlabel('bunch length, z = -c*t (m)')
plt.ylabel('Main+ Harmonic (MV)')
plt.title('Main + Harmonic voltage (MV) versus bunch length (m)',color='blue')
plt.plot(z,Main_plus_harmonic,color='green',lw=3,ls='-',label='Main+Harmonic')
plt.legend(loc=1)
```

```python
# Bunch length plot
c = 3e08
z= -c*t*100
plt.xlabel('bunch length, z = -c*t (cm)')
plt.ylabel('Main+ Harmonic (MV)')
plt.title('Main + Harmonic voltage (MV) versus bunch length (m)',color='blue')
plt.plot(z,Main_plus_harmonic,color='green',lw=3,ls='-',label='Main+Harmonic')
plt.legend(loc=1)
```
